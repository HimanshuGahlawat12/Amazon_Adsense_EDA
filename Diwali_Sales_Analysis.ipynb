**EDA - Exploratory Data Analysis**

We will use visuals to get a view of our data so that it is easy for management to take decisions

```python
df.columns
```

```python
ax = sns.countplot( x = 'Gender' , data = df)

for bars in ax.containers:
  ax.bar_label(bars)
```

**FEMALES HAVE MADE MORE PURCHASES THAN MALES**

```python
sales_gen = df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(x = 'Gender' , y = 'Amount' , data = sales_gen)
plt.title('Bar Plot of Amount by Gender')
```

**From above it is clear that women have more purchasing power than men**

**AGE**

```python
ax = sns.countplot(x = 'Age_group' , hue ='Gender' ,data = df)
plt.ylabel("Count of people in age group spending money")

for bars in ax.containers:
  ax.bar_label(bars)
```

**Customers between the age group of 26-35 have more purchasing power**


**Customer under 17 and 55+ have least purchasing power**

```python
sales_age = df.groupby(['Age_group'] , as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x = 'Age_group' , y = 'Amount' , data = sales_age)
plt.title('Amount Spent As Per Age Group')
```

**From it is clear that the 26 - 35 is the age group with most purchasing power**

```python
sales_state = df.groupby(['State'] , as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)

plt.figure(figsize=(12,5))
plt.title('Top 10 States With Maximum spend')
plt.xticks(rotation=75)
sns.barplot(x = 'State' , y = 'Amount' , data = sales_state)
```

**It is clear that Uttar Pradesh and Maharashtra are state with most purchasing power **

```python
sales_state = df.groupby(['State'] , as_index = False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)

plt.figure(figsize=(12,5))
plt.title('Top 10 States With Maximum Order')
plt.xticks(rotation=75)
sns.barplot(x = 'State' , y = 'Orders' , data = sales_state)
```

**It is clear that Uttar Pradesh and Maharashtra are state with most orders **

```python
ax = sns.countplot(data = df , x = 'Marital_Status')

sns.set(rc={'figure.figsize':(5,2)})
for x in ax.containers:
  ax.bar_label(x)
```

**Married women have maximu purchasing power**
